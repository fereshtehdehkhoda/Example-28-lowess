```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data=pd.read_csv("windxlsx.csv")
```

```python
data.info()
```

```python
plt.scatter(x=data['x'],y=data['y'],color='b')
plt.xlabel('wind speed')
plt.ylabel('mizane tolide bargh')
plt.title('wind speed vs. bargh')

#lowess
lowes_re=sm.nonparametric.lowess(data['y'],data['x'])
plt.plot(lowes_re[:,0],lowes_re[:,1],label='lowess line',color='green')
plt.legend()

#reg line
plt.plot(xp,yp,label='reg line',color='black')
plt.legend()


plt.show()
```

```python
xp=np.linspace(data['x'].min(),data['x'].max(),100)
yp=np.polyval(params,xp)
yp=np.polyval(params,6)
yp
```
